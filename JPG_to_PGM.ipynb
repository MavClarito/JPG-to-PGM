```python
import cv2
import numpy as np
import os
from google.colab import drive

drive.mount('/content/drive')
original_folder_path = '/content/drive/My Drive/Face_DataSet'
gray_folder_path = '/content/drive/My Drive/Gray_DataSet'

#Checking of file path
if not os.path.exists(original_folder_path):
    print("Original folder does not exist.")
elif not os.path.exists(gray_folder_path):
    print("Gray folder does not exist. ")

else:
    files = os.listdir(original_folder_path)
    for DataSet in files:
        if DataSet.endswith(('.jpg', '.jpeg')):
            image_path = os.path.join(original_folder_path, DataSet)
            image = cv2.imread(image_path)

            if image is not None:
                # Convert the image to grayscale
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                # Resize the image to 200x200 pixels
                resized_image = cv2.resize(gray_image, (200, 200))

                output_pgm_path = os.path.join(gray_folder_path, f'gray_{DataSet.split(".")[0]}.pgm')
                cv2.imwrite(output_pgm_path, resized_image, [int(cv2.IMWRITE_PXM_BINARY), 1])
                print(f"Converted {DataSet} to grayscale and saved as {output_pgm_path}")
            else:
                print(f"Unable to read {DataSet}")

print("Conversion complete.")
```
